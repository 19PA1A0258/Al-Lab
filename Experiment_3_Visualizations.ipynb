```
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

# i) Line Graphs

### Draw a line graph between SSC and intermediate/diploma percentage. Does it make sense? Explain when to use Line Graphs?

```
df=pd.read_csv('students_placement_data (3).csv')
```

```
plt.plot(df['SSC Percentage'], df['inter_Diploma_percentage'])
```

it is used identify the percentage between two attributes

# ii) Scatter Plots

### Draw a scatter plot between SSC and inter/diploma percentage. Do you find any correlation by just visual inspection? If yes, then what kind of correlation you can see? (+ve, -ve or No)

```
plt.scatter(df['SSC Percentage'], df['inter_Diploma_percentage'])
plt.xlabel('SSC score')
plt.ylabel('inter_diploma score')
plt.title('scatter plot between SSC and inter/diploma percentage')
```

It is clear that these two variables are positively correlated

# iii) Histograms

 ### Draw a histogram for B.Tech percentage. How many people have scores between 60-80% in B.tech. Tell your approximate answer by visual inspection?

```
bins=np.linspace(0,100,11)
print(bins)
plt.hist(df['B.Tech_percentage'],bins,width=9)
plt.xlabel('B.Tech score',fontsize=20)
```

 30-35 members have score (60-80)% in b.tech.

 # iv) Subplots

### Draw a scatter plot between     
####· intermediate and SSC 
####· Intermediate and B.Tech in the same graph.
###What's your observation?

```
fig,zy=plt.subplots()
zy.scatter(df['inter_Diploma_percentage'],df['SSC Percentage'],c='blue',marker='*',label='Inter-ssc')
zy.scatter(df['inter_Diploma_percentage'],df['B.Tech_percentage'],c='grey',marker='*',label='Inter-B.tech')
zy.set_xlabel('intermediate score',fontsize=17)
zy.legend(loc=0)
```

 The people who has good marks in inter is good in ssc but who get into b.tech marks are reduced as compared to ssc

# v) Join plots

 ### Draw a jointplot between intermediate and Backlogs in B.tech. What's your observation.

```
sns.jointplot(df['inter_Diploma_percentage'],df['Backlogs'],data=df)
```

 Those who got more marks in intermediate has less backlogs
